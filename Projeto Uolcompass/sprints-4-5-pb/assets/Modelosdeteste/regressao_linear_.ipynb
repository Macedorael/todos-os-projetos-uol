```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, accuracy_score

# Função para classificar o preço
def classify_price(price):
    if price <= 85:
        return 1
    elif 85 < price < 115:
        return 2
    else:
        return 3

# Carregar o dataset
base_hotel = pd.read_csv('/content/Hotel Reservations.csv')

# Criar a coluna de label
base_hotel['label_avg_price_per_room'] = base_hotel['avg_price_per_room'].apply(classify_price)

# Remover a coluna avg_price_per_room
base_hotel = base_hotel.drop(columns=['avg_price_per_room'])

# Colunas úteis para análise, incluindo a nova coluna 'label_avg_price_per_room'
features = ['lead_time', 'no_of_special_requests', 'no_of_adults',
            'no_of_children', 'type_of_meal_plan', 'arrival_month',
            'no_of_week_nights', 'room_type_reserved', 'label_avg_price_per_room']

# Seleciona as colunas especificadas no DataFrame base_hotel
X = base_hotel[features].copy()  # Usar .copy() para evitar o SettingWithCopyWarning

# Atribui a variável Y com os valores da coluna 'label_avg_price_per_room'
Y = X.pop('label_avg_price_per_room')

# Codificação das variáveis categóricas
label_encoders = {}
for column in X.select_dtypes(include=['object']).columns:
    le = LabelEncoder()
    X[column] = le.fit_transform(X[column])
    label_encoders[column] = le

# Converte X e Y para float32
X = X.astype('float32')
Y = Y.astype('float32')

# Dividir o conjunto de dados em treinamento e teste
X_treinamento, X_teste, Y_treinamento, Y_teste = train_test_split(X, Y, test_size=0.3, random_state=0)

# Treinar o modelo
model = LogisticRegression(max_iter=1000)
model.fit(X_treinamento, Y_treinamento)

# Fazer previsões
Y_pred = model.predict(X_teste)

# Avaliar o modelo
print("Accuracy Score:", accuracy_score(Y_teste, Y_pred))
print("Classification Report:\n", classification_report(Y_teste, Y_pred))
```
